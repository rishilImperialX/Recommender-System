```python
def convert_to_list(text):
    res = None
    if "," in text:
        res = text.split(',')
    else:
        res = text.split()
    return res
```

```python
import string
```

```python
pkt = string.punctuation
print(list(pkt))
pkt
```

```python
def remove_punctuation(text):
    for pk in list(pkt):
        if pk in text:
            text = text.replace(pk,'')
    return text
```

### stemming or lemmatization

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
def stem(text):
    res = []
    for i in text.split():
        res.append(lemmatizer.lemmatize(i))
    return " ".join(res)
```

```python
new_df['tags'][:10].values
```

```python
# new_df['tags_stem'] = new_df['tags'].apply(stem)
new_df['tags_lem'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags_lem']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names_out()[:100]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title,"\t","Score:",str(round(i[1] * 100,2))+'%') 
```

```python
recommend('Zoombies')
```
